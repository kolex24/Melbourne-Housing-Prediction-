```python
import datetime
def wrangle(filepath):
  df = pd.read_csv(filepath)
  df['Date'] = pd.to_datetime(df.Date, format='%d/%m/%Y')
  df['Car'].fillna(0, inplace=True)
  df['Car'].astype(int)
  df.drop(df[df['BuildingArea']==44515].index, inplace=True)
  df['BuildingArea'].fillna(df['BuildingArea'].median(), inplace=True)
  df.drop(df[df['YearBuilt']==1196].index, inplace=True)
  df['YearBuilt'].fillna(df.YearBuilt.median(), inplace=True)
  df['YearBuilt'].astype(int)
  df['CouncilArea'].fillna('No CounciArea', inplace=True)
  df.drop(df[df['Landsize']==433014].index, inplace=True)
  df['Year_Sold'] = df['Date'].dt.year
  df['Month_Sold'] = df['Date'].dt.month
  df['Day_Sold'] = df['Date'].dt.day
  df['HouseAge'] = df['Date'].dt.year - df['YearBuilt']
  df['Suburb__Median_Price'] = df.groupby('Suburb')['Price'].transform('median')
  df['Car_Parking'] = (df['Car'] > 0).astype(int)
  df['Region_Median_Price'] = df.groupby('Regionname')['Price'].transform('median')
  df['Room_Per_Bathroom'] = df['Rooms']/(df['Bathroom'] + 1)
  df.drop(['Propertycount','Address','Postcode', 'Date'], axis=1, inplace=True )
  return df
```

```python
df = wrangle('melb_data.csv')
```

### MELBOURNE HOUSING PREDICTION (USING DECISION TREE AND GRADIENT BOOSTING REGRESSION)



```python
x,y = df.drop('Price', axis=1), df['Price']
```

```python
x.describe() #Descriptive Statistics of Features
```

```python
#pip install category_encoders
```

```python
from category_encoders import OneHotEncoder
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor #importing My Model
from sklearn.metrics import accuracy_score,confusion_matrix #To Check Model Accuracy
from sklearn.model_selection import train_test_split #To Split Data into Train and Test
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score #Model Error
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2) #Splitting
```

```python
y_base = y_train.mean() *np.ones(len(y_train))
y_base[:10]
```

```python
mse_baseline = mean_squared_error(y_train, y_base)
mse_baseline
```

```python
rmse_baseline = np.sqrt(mse_baseline)
rmse_baseline
```

```python
#model = DecisionTreeRegressor(max_depth=5, random_state=42) #Defining MyModel
model = make_pipeline(
    OneHotEncoder(use_cat_names=True),
    DecisionTreeRegressor()
)
```

```python
model.fit(X_train,y_train) #Fitting MyModel
```

```python
y_pred = model.predict(X_train)
y_pred_test = model.predict(X_test)
```
